```python
alpha = np.linspace(0, 2*np.pi, 41)
```

```python
vectors = list(zip(np.cos(alpha), np.sin(alpha)))
```

```python
plot_vector(vectors)
```

```python
vectors[0]
```

```python
C.dot(vectors[0])
```

```python
newvectors = []

for i in range(len(vectors)):
    newvectors.append(C.dot(vectors[i]))
```

```python
plot_vector(newvectors)
```

```python
lengths = []
for i in range(len(newvectors)):
    lengths.append(np.linalg.norm(newvectors[i]))
```

```python
semi_major = max(lengths)
print(semi_major)
```

```python
semi_minor = min(lengths)
print(semi_minor)
```

```python
s = semi_major/np.sqrt(2)
```

```python
u1 = np.array([s, s])
print(u1)
```

```python
C_inv = np.linalg.inv(C)
v1 = C_inv.dot(u1)
```

```python
plot_vector([u1, v1])
```

The vector that landed on the semi-major axis was only scaled by the effect of $C$

$$ C \mathbf{v}_1 = s1 \mathbf{v}_1 $$
